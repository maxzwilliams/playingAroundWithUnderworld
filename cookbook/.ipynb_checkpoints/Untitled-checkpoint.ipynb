# Quickstart guide to underworld 3

Underworld 3 is a finite element method code that uses symolic repressentations for ease of model validation and construction

Heres how you import underworld3

```python
import underworld3 as uw
```

The fundamental object in underworld 3 is the mesh. This is the discritized domain on which we simulate. For example, here is a mesh on a rectangular domain $[0,1]\times[0,1]$ with cells of scale $\frac{1}{24}$

```python
mesh = uw.meshing.UnstructuredSimplexBox(minCoords=(0.0, 0.0), maxCoords=(1.0, 1.0), cellSize= 1.0 / (24), qdegree=5)
```

We can print out what our meshed domain looks like:

```python
## print out what the mesh looks like
if uw.mpi.size == 1:

    # plot the mesh
    import numpy as np
    import pyvista as pv
    import vtk

    pv.global_theme.background = "white"
    pv.global_theme.window_size = [750, 750]
    pv.global_theme.antialiasing = True
    pv.global_theme.jupyter_backend = "panel"
    pv.global_theme.smooth_shading = True

    mesh.vtk("tmp_mesh.vtk")
    pvmesh = pv.read("tmp_mesh.vtk")

    pl = pv.Plotter()

    pl.add_mesh(
        pvmesh,
        cmap="coolwarm",
        edge_color="Black",
        show_edges=True,
        use_transparency=False,
    )

    pl.show(cpos="xy")
```
